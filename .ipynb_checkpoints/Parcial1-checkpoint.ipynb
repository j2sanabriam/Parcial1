## Limpieza y preparación de datos

```python
# Crear una expresion lambda para transformar Float a Int
intToFloat = lambda x: int(x)
```

```python
# Aplicar lamda a todos los valores en la columna Date
df["construction year"] = df["construction year"].apply(intToFloat)
```

```python
# Aplicar lamda a todos los valores en la columna minimum nights
df["minimum nights"] = df["minimum nights"].apply(intToFloat)
```

```python
# Aplicar lamda a todos los valores en la columna availability 365
df["availability 365"] = df["availability 365"].apply(intToFloat)
```

```python
# Aplicar lamda a todos los valores en la columna number of reviews
df["number of reviews"] = df["number of reviews"].apply(intToFloat)
```

```python
# Aplicar lamda a todos los valores en la columna review rate number
df["review rate number"] = df["review rate number"].apply(intToFloat)
```

```python
# Crear una expresion lambda para transformar String (precio) a Float
priceToFloat = lambda x: float(str(x).replace("$", "").replace(",", "."))
```

```python
# Aplicar lamda a todos los valores en la columna price
df["price"] = df["price"].apply(priceToFloat)
```

```python
# Aplicar lamda a todos los valores en la columna price
df["service fee"] = df["service fee"].apply(priceToFloat)
```
